```python
import string
string.ascii_lowercase
from math import ceil
from itertools import cycle, islice, dropwhile
alph = list(string.ascii_lowercase)
```

```python
alphs = [None] * len(alph)
cycled = cycle(alph)
for i in range(0, len(alph)):
  skipped = dropwhile(lambda x: x != alph[i], cycled)
  alphs[i] = list(islice(skipped, None, len(alph)))
alphs
```

```python
def generate_words():
  curr_words = []
  for cycle_alph in alphs:
    for cycle_alph2 in alphs:
      for cycle_alph3 in alphs:
        for cycle_alph4 in alphs:
          curr_words.extend(list(zip(cycle_alph, cycle_alph2, cycle_alph3, cycle_alph4)))
  return curr_words
asdsa = generate_words()
three_chars_words = sorted(list(set([''.join(i) for i in asdsa])))
print(len(three_chars_words))
three_chars_words
```
